*se realiza la importacion del modulo integrado de Python Operador el cual nos permitira realizar funciones matematicas
*se crea una lista vacia llamada individual_features_df
*se crea un rango partiendo del indice cero dentro del subset de datos df_num y con el -1 eliminamos el ultimo valor el cual seria SalePrice
*se crea un subset de datos llamdo tmpDf a partir de df_num en el que pedimos el numero de valores en SalesPrice
*creamos el subset de datos en el que pedimos los valores siempre que sean diferentes a 0
y estos se incorporan al df indivuadual_features_df

Se crean correlaciones a partir de la Columna SalePrice y se imprimen de menor a mayor

```python
import operator

individual_features_df = []
for i in range(0, len(df_num.columns) - 1): 
    tmpDf = df_num[[df_num.columns[i], 'SalePrice']]
    tmpDf = tmpDf[tmpDf[df_num.columns[i]] != 0]
    individual_features_df.append(tmpDf)

all_correlations = {feature.columns[0]: feature.corr()['SalePrice'][0] for feature in individual_features_df}
all_correlations = sorted(all_correlations.items(), key=operator.itemgetter(1))
for (key, value) in all_correlations:
    print("{:>15}: {:>15}".format(key, value))
```

```python
golden_features_list = [key for key, value in all_correlations if abs(value) >= 0.5]
print("There is {} strongly correlated values with SalePrice:\n{}".format(len(golden_features_list), golden_features_list))
```

*Se crea una matriz de correlacion en la que se elimina la columna SalePrice y se toman las columnas a partir de la posicion 1, asi comenzando desde MS Sub Class
*se le asigna un tamaÃ±o de 12x10 a la figura
*y se utiliza el diseÃ±o de seaborn de mapa de calor para nuestra representacion de la matriz. en esta establecemos los valores  de la correlacion que queremos resaltar, ya sea si la correlaccion es igual o mayor a -0.4 o a 0.5. Con cmap definimos la escala de colores a utilizar (viridis, magma, turbo etc). Se define un valor minimo y un valor maximo. Se define el grosor de las lineas en nuestro grafico. annot=True quiere decir que si el Vmin y Vmax estan definidos imprimame el valor, si noestan definidos pues entonces no mostrara el numero. annot_kws define el tamaÃ±o del texto que aparezca dentro del mapa. square=True define la figura que queremos que tenga el mapa

```python
corr = df_num.drop('SalePrice', axis=1).corr() 
plt.figure(figsize=(12, 10))

sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```

*creamos la lista features_to_analyse a partir de la lista quantitative_features_list en la que se encuentren los valores de la lista golden_features_list

*Fig, ax es una abreviatura de fig = plt.figure() y ax = plt.subplot(). Por medio de esta funcion creamos un grafico a partir de uno existente, donde le pedimos que nos de la longitud lista features_to_analyse, y que lo divida entre tres y que adicionalmente redondee los valores.Le pedimos que despliegue 3 graficos en la misma linea de tamaÃ±o 18x12

*Pedimos enumerar al index dentro de fig.axes siempre que se cumpla la siguiente condicion:
si index es menor a la longitud de features_to_analyse entonces devuelvanos el ultimo valor. E imprimalo en un grafico de tipo replot de Seaborn donde X es igual a la informacion en features_to_analyse y Y equivale a SalePrice, se define de donde se esta tomando la informacion

```python
fig, ax = plt.subplots(round(len(features_to_analyse) / 3), 3, figsize = (18, 12))

for i, ax in enumerate(fig.axes):
    if i < len(features_to_analyse) - 1:
        sns.regplot(x=features_to_analyse[i],y='SalePrice', data=df[features_to_analyse], ax=ax)
```

Se crea la lista categorical_features donde le pedimos devolver los valores de quantitative_features_list excepto el ultimo item de la lista, pedimos con df.columns.tolist() que nos de el nombre de las columnas de data frame en formato de lista siempre que a no este en la lista quantitative_features_list o en df.columns.tolist/.

Se crea un subset dde datos a partir de los valores recien obtenidos y se imprimen.

```python
# quantitative_features_list[:-1] as the last column is SalePrice and we want to keep it
categorical_features = [a for a in quantitative_features_list[:-1] + df.columns.tolist() if (a not in quantitative_features_list[:-1]) or (a not in df.columns.tolist())]
df_categ = df[categorical_features]
df_categ.head()
```

se crea un subset dde ddatos en el que se selecciona todas aquellas columnas que no sean de tipo numerico y se imprimen en una lista 

```python
df_not_num = df_categ.select_dtypes(include = ['O'])
print('There is {} non numerical features including:\n{}'.format(len(df_not_num.columns), df_not_num.columns.tolist()))
```

Definimos la figura y tamaÃ±o a utilizar. Definimos la variable ax la cual contiene el grafico tipo box plot de Seaborn, donde X es igual a BsmtExposure y Y es SalePrice , del subset de datos llamado df_categ.

```python
plt.figure(figsize = (10, 6))
ax = sns.boxplot(x='BsmtExposure', y='SalePrice', data=df_categ)
plt.setp(ax.artists, alpha=.5, linewidth=2, edgecolor="k")
plt.xticks(rotation=45)
```

Se crean graficos a partir del subset de datos df_not_num donde se le asigna un tamano de 12x30, y 3 graficos por fila.

```python
fig, axes = plt.subplots(round(len(df_not_num.columns) / 3), 3, figsize=(12, 30))

for i, ax in enumerate(fig.axes):
    if i < len(df_not_num.columns):
        ax.set_xticklabels(ax.xaxis.get_majorticklabels(), rotation=45)
        sns.countplot(x=df_not_num.columns[i], alpha=0.7, data=df_not_num, ax=ax)

fig.tight_layout()
```
